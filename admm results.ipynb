```python
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
import pickle

with open('history_conv_layers.pickle') as f:
    history = pickle.load(f)
print history.keys()
keys = ['acc', 'loss', 'dual_to_consensus', 'distance_to_consensus']

for k in keys:
    figsize = (12,8)
    plt.figure(figsize=figsize)
    plt.plot(history[k])
    plt.grid()
    plt.title(k)
    if k == 'loss':
        plt.ylim(0,4)
```
